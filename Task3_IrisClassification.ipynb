**The
objective is to train a machine learning model that can learn from
these measurements and accurately classify the Iris flowers into
their respective species.**

Imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris_data = pd.read_csv("./IRIS.csv")
```

How the data looks

```python
iris_data.head()
```

```python
iris_data.shape
```

```python
iris_data['species'] = iris_data['species'].replace({"Iris-setosa":1, "Iris-versicolor":2, "Iris-virginica" : 3 })
```

```python
iris_data.head()
```

```python
iris_data.describe()
```

```python
y = iris_data['species']
x = iris_data.drop("species", axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, shuffle=True)
```

Training the model of Logistic Regression

```python
model = LogisticRegression()

model.fit(x_train, y_train)

pred = model.predict(x_test)
```

Checking accuracy on test dataset

```python
accuracy = accuracy_score(y_test, pred)
print('Accuracy of the model on test dataset: ', accuracy * 100)
```

Training the model of KNN

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
knn_pred = knn.predict(x_test)
```

Checking accuracy on test dataset

```python
accuracy = accuracy_score(y_test, knn_pred)
print('Accuracy of the model on test dataset: ', accuracy * 100)
```
