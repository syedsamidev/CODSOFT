**Use the Titanic dataset to build a model that predicts whether a
passenger on the Titanic survived or not.**

Imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_data = pd.read_csv("./tested.csv")
```

How the data looks

```python
print(titanic_data.shape)
titanic_data.head()
```

```python
titanic_data.describe()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data = titanic_data.drop(columns="Cabin", axis=1)
```

```python
titanic_data['Sex'] = titanic_data['Sex'].replace({'male': 0, 'female': 1})
titanic_data['Embarked'] = titanic_data['Embarked'].replace({'Q': 0, 'S': 1, 'C':3})
titanic_data.head()
```

```python
titanic_data = titanic_data.dropna().reset_index(drop=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.shape
```

```python
titanic_data = titanic_data.drop(columns=["Name","PassengerId","Fare","Ticket"],axis=1)
titanic_data.head()
```

```python
y = titanic_data["Survived"]
x = titanic_data.drop(["Survived"], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, shuffle=True)
```

Training the model on the train dataset

```python
model = LogisticRegression()

model.fit(x_train, y_train)

pred = model.predict(x_test)
```

Dtermining the accuracy of the model on the test dataset

```python
accuracy = accuracy_score(y_test, pred)
print('Accuracy of the model on test dataset: ', accuracy)
```
